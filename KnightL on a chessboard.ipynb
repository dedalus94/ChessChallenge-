```python

import math
import numpy as np 


"""
After some time analising the problem to solve, I have decided that I needed to create a function to call to make a move.
The function generates the 8 possible moves; it can simulates moves with different L shapes, as it was required by the assignment.
x and y can be changed to change the L shape and (a,b) is the starting point. 
Output corresponds to all possible moves from (a,b).
It is important to notice that an element of the list in ouput can be given as starting point to the function,
so that another move can be made." 
"""
def makeMove(a,b,x,y):
    
    moves= [[a+x,b+y],[a+x,b-y],[a-x,b+y],[a-x,b-y],[a+y,b+x],[a+y,b-x],[a-y,b+x],[a-y,b-x]]
    print("Possible moves from ({},{}) are:{}".format(a,b,moves))
    return moves 

"""We want to test the sequences of all possible moves, until we find the one that can lead to the final point (n,n).
Point that corresponds to the corner opposite to the starting point on the chessboard.
The goal is to find the minimum number of moves to reach it.

The first point of the 8 (**) generated as "available moves" -by calling makeMove- is always chosen first, the function 
then checks what points are available from that one, and it carries this way in a loop, discarding certain points
in the process and eventually going back to the second point generated at (**) if a dead end has been reached.

This function is called "PossibleMoves" but what it rather does is to generate a path: it calls itself in a loop
and at every iteration the chess piece advances, 
until the destination is one of the available moves generated by calling makeMove.
"""
final_count=0

def PossibleMoves(moves,x,y,n,count,previous_move):
    
    global final_count
    for move in moves:           #Because the function calls itself, and everytime a move has been made, 
                                 #the first thing to do is to check whether the move has reached the final point.
                                 #This does not interfere with the rest during the firt iteration, 
                                 #because the starting point is always (0,0) and therefore the if condition not fulfilled 
        if move==[n-1,n-1]:
            print("Result found!! {} moves were required".format(count))
            final_count=count
            return count
        
    
    for move in moves:
        
        if count>=n*2:   #This is a check to avoid loops
            final_count=-1
            return -1
        else:


            if move[0]>=n or move[0]<0 or move[1]>=n or move[1]<0: #This check discards all moves which are out of the 
                                                                   #perimeter of the chessboard
                                                                   #Because of the "pass", no  further points are generated
                pass

                                                                   
            else:                                                  

                                                                   
                moves=makeMove(move[0],move[1],x,y)

                if previous_move in moves:
        
                    moves.remove(previous_move)                  #This is to prevent some loops in
                                                                 #which the function generates the point of origin and then that point 
                                                                 #is chosen as the next move, leading to a back and forth loop.
                    count+=1  
                    previous_move=move

                    PossibleMoves(moves,x,y,n,count,previous_move)# If all the checks are passed, 
                                                                  # the point is not the destination and it is on the chessboard,
                                                                  #a new set of points is generated
                else:

                    count+=1                                      #Everytime the fuction is called, a move is made and count goes up by 1  
                    previous_move=move
                    PossibleMoves(moves,x,y,n,count,previous_move)


                return count
            final_count=-1 
            pass
    #print(count)
    
"""This function will call the previous ones. And it will find the shortest path from (0,0) to (n,n) 
with Knights that can move with different L shaped moves"""

def knightlOnAChessboard(n):
    a=0
    b=0
    
    store=[]
    for x in range(1,n): #fissa x 
        row=[]
        store.append(row)
        for y in range(1,n): #itera su y con x fissata 
            print("This Knight can make the move: ({},{})".format(x,y))
            
                
            
            
            moves=makeMove(a,b,x,y)
            PossibleMoves(moves,x,y,n,1,[])
            print(final_count)
            row.append(final_count) 
        print(row)
    print("store:".format(store))
    return store          
   
                    
knightlOnAChessboard(5) 
```
